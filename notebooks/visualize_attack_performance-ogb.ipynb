```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../reports/initial_eval-ogb.csv')
```

```python
df.loc[df.attack=='StructackBothEndsGreedy','attack']='Ours'
```

```python
df = df[df.attack.isin('Clean DICE Random Ours Metattack'.split())]
```

```python
g = sns.FacetGrid(df[df.attack!='Clean'],col='dataset',sharey=False,legend_out=True)
g.map_dataframe(sns.lineplot,x='perturbation_rate',y='acc',hue='attack',marker='o')
g.add_legend()
plt.savefig('../reports/figures/acc-ogb.pdf',bbox_inches='tight')
```

```python
g = sns.FacetGrid(df[df.attack!='Metattack'],col='dataset',sharey=False,legend_out=True)
g.map(sns.lineplot,'perturbation_rate','elapsed','attack')
g.add_legend()
```
