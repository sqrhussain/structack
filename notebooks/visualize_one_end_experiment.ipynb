```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../reports/eval/one_end.csv')
```

```python
df2 = pd.read_csv('../reports/eval/one_end-ptb.csv')
```

```python
df = pd.concat([df,df2],ignore_index=True)
```

```python
for dataset in df.dataset.unique():
    cdf = df[df.dataset == dataset]
    plt.figure()
    g = sns.lineplot(data=cdf,x='frm', y='acc', hue='perturbation_rate', marker='o')
    g.yaxis.set_major_formatter(FormatStrFormatter('%.4f'))
    plt.xlabel('Degree percentile')
    plt.ylabel('Accuracy')
    plt.legend(title='Perturbation rate')
    plt.title(dataset)
    plt.savefig(f'../reports/figures/edge-single-end-degree/{dataset}.pdf',bbox_inches='tight')
    plt.show()
    
    
```
