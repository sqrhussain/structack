```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
sns.set_style('whitegrid')
```

```python
df_unnoticeability = pd.read_csv('../reports/eval/degre_noticeability.csv')
```

```python
df_unnoticeability['alpha_abs_difference'] = abs(df_unnoticeability.alpha_modified - df_unnoticeability.alpha_original)
df_unnoticeability['ccoefs_abs_difference'] = abs(df_unnoticeability.mean_clustering_coef_modified - df_unnoticeability.mean_clustering_coef_orig)
```

```python
for dataset in df_unnoticeability.dataset.unique():
    cdf = df_unnoticeability[df_unnoticeability.dataset == dataset]
    plt.figure(figsize=(12,8))
    g = sns.lineplot(data=cdf,x='perturbation_rate', y='alpha_abs_difference', hue='attack', marker='o')
    g.yaxis.set_major_formatter(FormatStrFormatter('%.4f'))
    plt.xlabel('Perturbation rate')
    plt.ylabel('Alpha exponent absolute difference')
    plt.legend(title='Perturbation rate')
    plt.title(dataset)
#     plt.savefig(f'../reports/figures/edge-single-end-degree/{dataset}.pdf',bbox_inches='tight')
    plt.show()
```
